```python
x = np.linspace(0, 10, 100)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.show()
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x,np.sin(x),color='r')
plt.plot(x,np.cos(x),color='y')
plt.show()
```

```python
plt.plot(x, np.sin(x - 0), color='blue')        # color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk) 
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1 
plt.plot(x, np.sin(x - 5), color='chartreuse'); # HTML color names
plt.show()
```

```python
plt.plot(x, x + 0, linestyle='solid')  
plt.plot(x, x + 1, linestyle='dashed')   
plt.plot(x, x + 2, linestyle='dashdot')   
plt.plot(x, x + 3, linestyle='dotted')
plt.show()
```

```python
# Short codes for linestyles
plt.plot(x, x + 4, linestyle='-')  # solid  
plt.plot(x, x + 5, linestyle='--') # dashed  
plt.plot(x, x + 6, linestyle='-.') # dashdot   
plt.plot(x, x + 7, linestyle=':')  # dotted
```

```python
# Short codes for linestyles
plt.plot(x, x + 4, linestyle='-',linewidth=4)  # solid  
plt.plot(x, x + 5, linestyle='--') # dashed  
plt.plot(x, x + 6, linestyle='-.') # dashdot   
plt.plot(x, x + 7, linestyle=':')  # dotted
plt.show()
```

```python
# (rows, columns, indices)
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5, 2, 4, 2, 1, 4, 5, 2]
plt.subplot(2 , 2 , 1)# create a subplot grid of 1x2 and use 1
plt.plot(x, y , 'r^')
plt.subplot(2 , 2 , 2)
plt.plot(x, y, 'b*-')
plt.subplot(2 , 2 , 3)
plt.subplot(2 , 2 , 4)
plt.show()
```

```python
# (rows, columns, indices)
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5, 2, 4, 2, 1, 4, 5, 2]
plt.subplot(4 , 2 , 1)# create a subplot grid of 1x2 and use 1
plt.plot(x, y , 'r')
plt.subplot(4 , 2 , 2)
plt.plot(x, y, 'b*-')
plt.subplot(4 , 2 , 3)
plt.plot(x, y, 'b*-')
plt.subplot(4 , 2 , 4)
plt.plot(x, y, 'b*-')
plt.subplot(4 , 2 , 5)
plt.plot(x, y, 'b*-')
plt.subplot(4 , 2 , 6)
plt.plot(x, y, 'b*-')
plt.subplot(4 , 2 , 7)
plt.plot(x, y, 'b*-')
plt.subplot(4 , 2 , 8)
plt.plot(x, y, 'b*-')
plt.show()
```
